# ¿Por qué usar seaborn?

- Es una librería "wrapper" de matplotlib
- Mucho más cómoda de usar.
- Estética más trabajada.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 500)
y = np.cumsum(np.random.randn(500, 6), 0)
```

```python
plt.plot(x, y)
```

```python
# Si queremos meter una leyenda...
plt.plot(x, y)
plt.legend('ABCDEF')
```

```python
# Si lo queréis hacer más feo... (estilo clásico)
plt.style.use('classic')
plt.plot(x, y)
plt.legend('ABCDEF')
```
